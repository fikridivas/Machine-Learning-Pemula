```python
from sklearn import preprocessing
```

```python
data = [[12000000, 33], [35000000, 45], [4000000, 23], [6500000, 26], [9000000, 29]]
```

```python
# buat objek scaler dan panggil fungsi fit
scaler = preprocessing.StandardScaler().fit(data) # menghitung rata" dan deviasi standar dari setiap kolom atribut
```

```python
# aplikasikan standar scaler
data = scaler.transform(data)
data
```
