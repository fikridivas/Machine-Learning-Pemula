```python
import pandas as pd
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
# ubah nama kolom
df = df.rename(columns={'Gender': 'gender', 'Age': 'age',
                        'Annual Income (k$)': 'annual_income',
                        'Spending Score (1-100)': 'spending_score'})
```

```python
# ubah data kategorik mmenjadi data numerik
df['gender'].replace(['Female', 'Male'], [0,1], inplace=True)
```

```python
df.head()
```

```python
# import k=means
from sklearn.cluster import KMeans
```

```python
# menghilangkan kolom customer id dan gender
X = df.drop(['CustomerID', 'gender'], axis=1)
```

```python
#membuat list yang berisi inertia, Di sini kita akan membuat list yang berisi inersia dari nilai K antara 1 sampai 11.
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
 
# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# kita bisa melihat bahwa elbow berada di nilai K sama dengan 5, 
# di mana penurunan inersia tidak lagi signifikan setelah nilai K sama dengan 5
```

```python
# Terakhir kita bisa melatih kembali K-Means dengan jumlah K yang didapat dari metode Elbow. 
```

```python
# membuat objek k-means
km5 = KMeans(n_clusters=5).fit(X)

# menambah kolom label pada dataset
X['Labels'] = km5.labels_

# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(X['annual_income'], X['spending_score'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```
