```python
import sklearn
from sklearn import datasets
 
# Load iris dataset
iris = datasets.load_iris()
```

```python
# mendefinisikan atribut dan label pada dataset
x=iris.data
y=iris.target
```

```python
# membuat model dengan decision tree
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
# mengevaluasi performa model dengan cross_val_score
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, x, y, cv=5) # cv merupakan jumlah fold yang akan dipakai pada cross validation
```

```python
# melihat akurasi pengujian pada setiap fold dari dataset
scores
```

```python
# Secara umum jika hasil dari pengujian tiap fold pada cross validation memiliki 
# nilai yang bervariasi dari 0.85 sampai 0.99, maka model tersebut dapat dikatakan baik.
```
