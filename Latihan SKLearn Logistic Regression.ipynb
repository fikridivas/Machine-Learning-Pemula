```python
import pandas as pd
```

```python
# baca dataset dari file csv
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head()
```

```python
# melihat data apakah ada nilai yang kosong menggunakan info()
data.info()
```

```python
# drop kolom User ID karena tidak penting
data = data.drop(columns=['User ID'])
```

```python
# lakukan proses one-hot-encoding karena masih ada bentuk kategorikal
data = pd.get_dummies(data)
data
```

```python
# pisahkan atribut label dan kolom
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']
X = data[predictions]
y = data['Purchased']
```

```python
# lakukan proses normalisasi sebelum membagi datanya
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```python
# membagi data menggunakan train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
# buat model logistic regression
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
# menguji akurasi model
from sklearn.preprocessing import StandardScaler

model.score(X_test,y_test)
```
