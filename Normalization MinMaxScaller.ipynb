```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = [[12000000, 33], [35000000, 45], [4000000, 23], [6500000, 26], [9000000, 29]]
```

```python
data
```

```python
# membuat objek minmaxscaler
scaler = MinMaxScaler()
scaler.fit(data) # menghitung nilai min dan max setiap kolom
```

```python
#  aplikasikan scaler pada data dengan menggunakan transform
print(scaler.transform(data))
```
